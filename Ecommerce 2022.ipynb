```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv(r"c:\Users\Hadi\Project Data\dataset-ecommerce.csv")
data
```

```python
data.info()
```

```python
#Ubah bagian Sales - float64 menjadi int64
data['Sales']=data['Sales'].astype('int64')
data.info()
```

```python
data['Sales'].unique()
```

Sekarang saya lakukan EDA dengan beberapa grafik yang keren :) Pertama, kita akan melihat bagaimana muatan didistribusikan sesuai dengan faktor yang diberikan.

```python
sns.set(style='whitegrid')
ax = plt.subplots(1,1, figsize=(12, 8))
ax = sns.distplot(data['Sales'], kde = True, color = 'y')
plt.title('Penjualan');
```

```python
ax = plt.subplots(1, 1, figsize=(12, 8))
ax = sns.distplot(np.log10(data['Sales']), kde = True, color = 'r' )
```

Sekarang mari kita lihat kategori produk yang paling tertinggi.

```python
ax = plt.subplots(1, 1, figsize=(12, 8))
ax = sns.barplot(x='Product Category', y='Sales', data=data, palette='cool')
```

dan sekarang mari kita lihat produk yang paling terlaris.

```python
data['Product'].unique()
```

```python
plt.figure(figsize=(13,30))
color=['lime', 'darkgreen', 'green', 'forestgreen', 'limegreen', 'mediumseagreen', 'mediumspringgreen', 'mediumaquamarine', 'lightseagreen', 'turquoise']
plt.pie(data['Product'].value_counts(), labels=data['Product'].value_counts().index, autopct='%1.1f%%', colors=color);
plt.title('Product');
```

Jadi secara keseluruhan, kategori produk tertinggi ada di bagian Fashion dan terendah ada di bagian Home&Adventure. Dengan mempertimbangkan faktor-faktor tertentu.

```python
ax = plt.subplots(1,1, figsize=(12,8))
ax = sns.barplot(x = 'Product Category', y = 'Sales',
                 hue='Region', data=data, palette='Reds_r')
```

ini hasil setiap bulan

```python
ax = plt.subplots(1, 1, figsize=(10, 10))
ax = sns.violinplot(x = 'Product Category', y = 'Sales', data=data,
                 orient='v', hue='Months', palette='magma')
```
